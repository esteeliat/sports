```python
# Open the VCF file
vcf_file = "cohort_brca17.vcf"  # Replace with your VCF file path
vcf = pysam.VariantFile(vcf_file)

# Dictionary to store counts for each variant ID
alt_allele_counts = {}

# Iterate through each record in the VCF
for record in vcf:
    # Count samples with at least one alternative allele
    count = sum(1 for sample in record.samples.values() 
                if sample['GT'] is not None and sum(allele for allele in sample['GT'] if allele is not None) > 0)

    
    # Store count using the variant ID
    alt_allele_counts[record.id] = count

# Print results
for variant_id, count in alt_allele_counts.items():
    print(f"{variant_id}: {count}")
```
